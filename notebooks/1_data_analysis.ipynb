## 1. Dataset description

```python
# Open dataset file 'MX4_Dataset.csv'
dataset = pd.read_csv(os.path.join(path_to_data, 'MX4_Dataset.csv'), index_col=[0])
dataset.head(10)
```

In each row there is one object corresponding one crystal structure of hybrid or all-inorganic group 15 metal halide with {MX<sub>4</sub>}<sup>–</sup> anion (M is metal: Bi<sup>3+</sup>, Sb<sup>3+</sup>; X is halogen: I<sup>–</sup>, Br<sup>–</sup>, Cl<sup>–</sup>).

Below information description contained in the columns:

* **Refcode** - crystallography database identifier (CSD, if the structure from ICSD or COD Refcode contain prefix “**ICSD_**” or “**COD_**”, respectively);

* **DOI**, **Authors**, **Journal**, **Journal year**, **Journal volume**, **Journal page or number** - information about publication in which the crystal structure is reported;

* **Cation SMILES** – SMILES representation of cation;

* **Cation substructure SMILES** – SMILES representation of cation substructure: cation including solvated and cocrystal molecules;

* **Space group**, **Int Tables number**, **Temperature**, **a**, **b**, **c**, **alpha**, **beta**, **gamma**, **V**, **Z**, **R factor** - crystallographic data about the crystal structure symmetry and the X-ray diffraction experiment;

* **MX4 type** - type of edge-shared $\{MX_{4}\}^{-}$ type: **alpha**, **beta**, **gamma**, (**See bellow I**), **no data** - *no structural data*;

* **M** - metal composition: **Bi**, **Sb** or **Bi Sb** for mixed compositions;

* **X** - halogen composition: **I**, **Br** or **Cl**, for mixed compositions – **Cl Br** or **Br I**;

* **X_t1**, **X_t2**, **X_d1**, **X_d2**, **X_d3**, **X_d4** - M–X distances labeled according to **II**, Å;

* **t1_t2**, **t1_d1**, **t1_d2**, **t1_d3**, **t1_d4**, **t2_d1**, **t2_d2**, **t2_d3**, **t2_d4**, **d1_d2**, **d1_d3**, **d1_d4**, **d2_d3**, **d2_d4**, **d3_d4** - X–M–X angles labeled according to **II**, °;

* **d_average** – average M–X distance, Å.

* **delta_d** - mean quadratic elongation ${\Delta}d$:

${\Delta}d=\frac{1}{6}\displaystyle\sum_{i=1}^{6}\left(\frac{d_{i}-\bar{d}}{\bar{d}}\right)^{2}$

where $d$ is individual M-X distance, $\bar{d}$ is the average M–X distance;

* **sigma_2** - angle variance $\sigma^{2}$:

 $\sigma^{2}=\frac{1}{11}\displaystyle\sum_{i=1}^{12}\left(\alpha_{i}-90\right)^{2}$
 
where $\alpha$ is individual <i>cis</i>-angle X-M-X;

* **X...X contacts** – space-separated string of X···X contacts distances (≤ 2·r<sub>VdW</sub>(X) + 1 Å) originated from one octahedron;

* **WdV contacts** – space-separated string of X···X contacts distances that equal or less than sum of van der Waals radii;

* **N_XX** – the number of X···X contacts (≤ 2·r<sub>VdW</sub>(X) + 1 Å) originated form one \[MX<sub>6</sub>\] octahedron;

* **N_VdW** – the number of van der Waals X···X contacts originated form one \[MX<sub>6</sub>\] octahedron;

* **XX_min** – minimal distance between halogen atom in neighboring chin anions;

* **X_average** – average X···X contact distance (≤ 2·r<sub>VdW</sub>(X) + 1 Å) originated form one \[MX<sub>6</sub>\] octahedron;

* **VdW_average** – average length of van der Waals X···X contacts originated form one \[MX<sub>6</sub>\] octahedron;

* **Type of bandgap** - bandgap type: **direct**, **indirect**, **XCCT** (charge transfer from halometallate anion to Cation), *CC* (Cation centrated);

* **Bandgap** - band gap value, eV;

* **Bandgap estimation comments** – comments to correctness of band gap estimation; not empty for incorrect determination.

## 3. Prepare for ML development

```python
# Generate d(X···X)/2·rVsW(X) desciptors
data['XXmin_2r'] = (data['XX_min']/composition_aid['X'].replace({'I': 4.2, 'Br I': (4.2+3.7)/2, 'Br': 3.7, 'Cl': 3.6})).astype(np.float64)
data['XXaver_2r'] = (data['XX_average']/composition_aid['X'].replace({'I': 4.2, 'Br I': (4.2+3.7)/2, 'Br': 3.7, 'Cl': 3.6})).astype(np.float64)
data['VdW_2r'] = (data['VdW_average']/composition_aid['X'].replace({'I': 4.2, 'Br I': (4.2+3.7)/2, 'Br': 3.7, 'Cl': 3.6})).astype(np.float64)
data = data[data.columns[:-4].to_list() + data.columns[-3:].to_list()+['Bandgap']]
# Generate composition descriptors
data['Composition'] = composition_aid['M'] + '_' + composition_aid['X']
data['M'] = data['Composition'].apply(lambda x: {'Bi': 0, 'Bi Sb': 0.5, 'Sb': 1}[x.split('_')[0]])
data['I'] = data['Composition'].apply(lambda x: {'I': 1, 'Br I': 0.5, 'Br': 0, 'Cl': 0}[x.split('_')[1]]) 
data['Br'] = data['Composition'].apply(lambda x: {'I': 0, 'Br I': 0.5, 'Br': 1, 'Cl': 0}[x.split('_')[1]])
data['Cl'] = data['Composition'].apply(lambda x: {'I': 0, 'Br I': 0, 'Br': 0, 'Cl': 1}[x.split('_')[1]])
data = data[['Refcode', 'Composition'] + data.columns[1:-6].to_list()+['M', 'I', 'Br', 'Cl', 'Bandgap']]
```

```python
data.to_csv(os.path.join(path_to_data, 'task.csv'))
data.to_excel(os.path.join(path_to_data, 'task.xlsx'))
```

```python
# Pearson correlation coefficients
p1, p2 = np.corrcoef(data['Bandgap'], data['delta_d'])[0, 1], np.corrcoef(data['Bandgap'], data['XXaver_2r'])[0, 1]
```
